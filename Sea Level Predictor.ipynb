```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Step 1: Import the data
df = pd.read_csv("epa-sea-level.csv")
```

```python
# Step 2: Create a scatter plot
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
```

```python
# Step 3: Create a line of best fit for the entire data range
slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])

# Plot the line of best fit
years_extended = range(1880, 2051)  # Extend the years to 2050
plt.plot(years_extended, [slope * year + intercept for year in years_extended], label="Best fit line (1880-2050)", color='red')
```

```python
# Step 4: Create a line of best fit for data from the year 2000 onwards
df_2000 = df[df['Year'] >= 2000]
slope_2000, intercept_2000, _, _, _ = linregress(df_2000['Year'], df_2000['CSIRO Adjusted Sea Level'])

# Plot the line of best fit from 2000 onwards
plt.plot(years_extended, [slope_2000 * year + intercept_2000 for year in years_extended], label="Best fit line (2000-2050)", color='green')
```

```python
# Step 5: Set labels and title
plt.title("Rise in Sea Level")
plt.xlabel("Year")
plt.ylabel("Sea Level (inches)")

# Step 7: Show the plot
plt.show()
```
